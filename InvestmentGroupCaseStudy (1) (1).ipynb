CheckPoint:1 (Data Cleaning)

```python
# Write your code for importing the csv file here

# In order to import the companies and rounds2 file, we'll be using 'read_csv()' method avaiable in 'Pandas' package and output data will be stored in 'companies' and 'rounds2' dataframe variable.
# NOTE:'companies.txt' & 'rounds2.csv' data input file, should be placed in same file path from where Jupyter Notebook is getting loaded.
companies = pd.read_csv('C:/Users/Abinash/Downloads/companies.txt', sep='\t', encoding = "ISO-8859-1")
rounds2 = pd.read_csv('C:/Users/Abinash/Desktop/rounds2.csv', encoding = "ISO-8859-1")
```

```python
print(companies.shape)
print(companies.info())
```

```python
# Dropping the un-necessary rows city, founded_at, region from the companies data frame
companies = companies.drop(['region','city','founded_at'],axis = 1)
companies
```

```python
#As for analysis we do not require homepage_url,state_code so not handling there null values as we wont use them at all.
#handling the data frame based on the other null values
#NOTE: There is a company which do not have a name, As we do not need company name for analysis letting it be
#companies = companies.dropna(subset=['name'])
#We cannot fill the country code with some average value or 0 so removing the rows which has null for the country code
companies = companies.dropna(subset=['country_code'])
##We cannot fill the category list with some average value or 0 so removing the rows which has null for the category list
companies = companies.dropna(subset=['category_list'])
print(companies.info())
```

```python
#unique companies present for companies data frame is
companies['permalink'].nunique()
```

```python
rounds2.info()
```

```python
#As for analysis we do not require funding_round_code so not handling the null values as we wont use them at all.
#As raised amount usd is a primary column for analysis , so removing the rows which do not have the values.
#As the raised amount in usd rage is very high taking the mean would be in apropiate and filling it with zeros will
#aslo be in apropiate
rounds2 = rounds2.dropna(subset=['raised_amount_usd'])
print(rounds2.info())
```

```python
#check for the uniquness
print(rounds2['company_permalink'].nunique())
```

```python
#Adding rows to both the data frame where value is in lower case for merging. As merge is case sensitive.
rounds2['company_permalink_lower'] = rounds2['company_permalink'].str.lower()
companies['permalink_lower'] = companies['permalink'].str.lower()
master_frame = rounds2.merge(companies , how= 'inner', left_on=['company_permalink_lower'], right_on=['permalink_lower'])
master_frame.info()
#After performing all of the operation we still have 76 percentage of the data set for analysis.
```

```python
master_frame
```

```python
master_frame = master_frame.drop(['company_permalink_lower','permalink','permalink_lower'],axis = 1)
master_frame.info()
```

```python
#taking care of the encoding issue for company_permalink column in master data frame
master_frame.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
```

```python
#Converting the raised_amount_usd from dollars to million dollars for ease of calculation
master_frame['raised_amount_usd'] = master_frame['raised_amount_usd']/1000000
print(master_frame.info())
master_frame
```

Checkpoint:2(Funding Type Analysis)

```python
#creating a group by object based on funding round type
master_fund_cound_type = master_frame.groupby('funding_round_type')
```

```python
#using the group by object on raised amount usd to get the average money spent at each sector.
avg_value = pd.DataFrame(master_fund_cound_type['raised_amount_usd'].mean())
avg_value
```

Checkpoint:3(Country Analysis)

```python
#Filtering the data set based of the funding round type
filter_value = master_frame['funding_round_type'] == 'venture'
#creating a new data set based of the funding round type
filtered_master_frame = master_frame[filter_value]
filtered_master_frame
```

```python
master_country_code = filtered_master_frame.groupby('country_code')
```

```python
total_fund = pd.DataFrame(master_country_code['raised_amount_usd'].sum().sort_values(ascending = False))
total_fund
```

```python
top9 = total_fund.iloc[0:9,:]
print(top9.info())
top9
```
